```python
import pandas as pd
import numpy as np
import json
import yaml
import joblib
import h5py

import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, confusion_matrix

from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC

# =============================
# PATHS
# =============================

DATA_PATH = r"C:\Users\NXTWAVE\Downloads\Water Infrastructure Priority Prediction\RS_Session_259_AU_1152_C.csv"
OUTPUT_DIR = r"C:\Users\NXTWAVE\Downloads\Water Infrastructure Priority Prediction"

# =============================
# LOAD DATA
# =============================

df = pd.read_csv(DATA_PATH)

# =============================
# RENAME COLUMNS
# =============================

df = df.rename(columns={
"Name of District": "District",
"Number of Urban Local Bodies (ULBs) in the district": "ULBs",
"Total households in the ULBs in 2021": "Total_Households",
"Number of households in these ULBs with Tap connections": "Tap_Connections",
"Number of households in these ULBs where Tap connections are yet to be provided": "No_Tap"
})

# =============================
# FEATURE ENGINEERING
# =============================

df["Coverage_Percentage"] = (
    df["Tap_Connections"] / df["Total_Households"] * 100
)

df["Gap_Households"] = (
    df["Total_Households"] - df["Tap_Connections"]
)

df["ULB_Density"] = (
    df["ULBs"] / df["Total_Households"]
)

# =============================
# CREATE PRIORITY LABEL
# =============================

df["Priority"] = pd.qcut(df["Coverage_Percentage"], 3, labels=[0,1,2])

# =============================
# FEATURES
# =============================

X = df[[
"Coverage_Percentage",
"Gap_Households",
"ULB_Density"
]]

y = df["Priority"]

# =============================
# SPLIT DATA
# =============================

X_train, X_test, y_train, y_test = train_test_split(
X, y, test_size=0.3, random_state=42
)

# =============================
# SCALING
# =============================

scaler = StandardScaler()

X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# =============================
# MODELS
# =============================

models = {
"Logistic Regression": LogisticRegression(),
"Decision Tree": DecisionTreeClassifier(),
"Random Forest": RandomForestClassifier(),
"SVM": SVC()
}

accuracies = {}

# =============================
# TRAIN MODELS
# =============================

for name, model in models.items():
    
    model.fit(X_train, y_train)
    
    pred = model.predict(X_test)
    
    acc = accuracy_score(y_test, pred)
    
    accuracies[name] = acc
    
    print(name, "Accuracy:", acc)

# =============================
# ACCURACY COMPARISON GRAPH
# =============================

plt.figure()

plt.bar(accuracies.keys(), accuracies.values())

plt.title("Model Accuracy Comparison")

plt.xticks(rotation=30)

plt.ylabel("Accuracy")

plt.savefig(OUTPUT_DIR + r"\accuracy_comparison.png")

plt.show()

# =============================
# HEATMAP
# =============================

corr = df[[
"Coverage_Percentage",
"Gap_Households",
"ULB_Density",
"Total_Households"
]].corr()

plt.figure()

sns.heatmap(corr, annot=True, cmap="coolwarm")

plt.title("Feature Correlation Heatmap")

plt.savefig(OUTPUT_DIR + r"\feature_heatmap.png")

plt.show()

# =============================
# TRAIN BEST MODEL
# =============================

best_model = RandomForestClassifier()

best_model.fit(X_train, y_train)

predictions = best_model.predict(X_test)

# =============================
# CONFUSION MATRIX
# =============================

cm = confusion_matrix(y_test, predictions)

plt.figure()

sns.heatmap(cm, annot=True)

plt.title("Confusion Matrix")

plt.savefig(OUTPUT_DIR + r"\confusion_matrix.png")

plt.show()

# =============================
# PREDICTION GRAPH
# =============================

plt.figure()

plt.scatter(y_test, predictions)

plt.xlabel("Actual")

plt.ylabel("Predicted")

plt.title("Prediction Graph")

plt.savefig(OUTPUT_DIR + r"\prediction_graph.png")

plt.show()

# =============================
# SAVE PREDICTION CSV
# =============================

result_df = pd.DataFrame({
"Actual": y_test,
"Predicted": predictions
})

prediction_csv = OUTPUT_DIR + r"\predictions.csv"

result_df.to_csv(prediction_csv, index=False)

print("Prediction CSV saved")

# =============================
# RESULT GRAPH
# =============================

priority_counts = df["Priority"].value_counts()

plt.figure()

priority_counts.plot(kind="bar")

plt.title("Priority Distribution")

plt.savefig(OUTPUT_DIR + r"\priority_distribution.png")

plt.show()

# =============================
# SAVE RESULT CSV
# =============================

result_file = OUTPUT_DIR + r"\district_results.csv"

df.to_csv(result_file, index=False)

print("Result CSV saved")

# =============================
# SAVE MODEL
# =============================

joblib.dump(best_model, OUTPUT_DIR + r"\water_model.pkl")

# =============================
# SAVE H5 MODEL
# =============================

with h5py.File(OUTPUT_DIR + r"\water_model.h5", "w") as f:
    f.create_dataset("predictions", data=np.array(predictions))

# =============================
# SAVE YAML
# =============================

metadata = {
"project": "Water Infrastructure Priority Prediction",
"best_model": "Random Forest",
"features": [
"Coverage_Percentage",
"Gap_Households",
"ULB_Density"
]
}

with open(OUTPUT_DIR + r"\model_metadata.yaml", "w") as f:
    yaml.dump(metadata, f)

# =============================
# SAVE JSON
# =============================

json_file = OUTPUT_DIR + r"\results.json"

with open(json_file, "w") as f:
    json.dump(result_df.to_dict(), f)

print("\nALL RESULTS SAVED SUCCESSFULLY")
```
