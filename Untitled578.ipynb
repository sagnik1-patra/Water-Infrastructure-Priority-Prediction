```python
import pandas as pd
import numpy as np
import json
import yaml
import joblib
import h5py

import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, confusion_matrix

# =============================
# PATHS
# =============================

DATA_PATH = r"C:\Users\NXTWAVE\Downloads\Water Infrastructure Priority Prediction\RS_Session_259_AU_1152_C.csv"
OUTPUT_DIR = r"C:\Users\NXTWAVE\Downloads\Water Infrastructure Priority Prediction"

# =============================
# LOAD DATA
# =============================

df = pd.read_csv(DATA_PATH)

df = df.rename(columns={
"Name of District": "District",
"Number of Urban Local Bodies (ULBs) in the district": "ULBs",
"Total households in the ULBs in 2021": "Total_Households",
"Number of households in these ULBs with Tap connections": "Tap_Connections",
"Number of households in these ULBs where Tap connections are yet to be provided": "No_Tap"
})

# =============================
# FEATURE ENGINEERING
# =============================

df["Coverage_Percentage"] = df["Tap_Connections"] / df["Total_Households"] * 100
df["Gap_Households"] = df["Total_Households"] - df["Tap_Connections"]
df["ULB_Density"] = df["ULBs"] / df["Total_Households"]

df["Priority"] = pd.qcut(df["Coverage_Percentage"], 3, labels=[0,1,2])

features = df[["Coverage_Percentage","Gap_Households","ULB_Density"]]
target = df["Priority"]

# =============================
# TRAIN TEST SPLIT
# =============================

X_train, X_test, y_train, y_test = train_test_split(
features, target, test_size=0.3, random_state=42
)

scaler = StandardScaler()

X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# =============================
# PSO OPTIMIZATION
# =============================

def pso_optimize(num_particles=20, iterations=40):

    dim = X_train.shape[1]

    particles = np.random.rand(num_particles, dim)
    velocities = np.zeros((num_particles, dim))

    pbest = particles.copy()
    pbest_scores = np.zeros(num_particles)

    gbest = particles[0]
    gbest_score = 0

    for _ in range(iterations):

        for i in range(num_particles):

            weights = particles[i]

            pred = np.dot(X_train, weights)
            pred = (pred > np.median(pred)).astype(int)

            score = np.mean(pred == y_train)

            if score > pbest_scores[i]:
                pbest_scores[i] = score
                pbest[i] = particles[i]

            if score > gbest_score:
                gbest_score = score
                gbest = particles[i]

        for i in range(num_particles):

            velocities[i] = (
                0.5 * velocities[i]
                + 1.5 * np.random.rand() * (pbest[i] - particles[i])
                + 1.5 * np.random.rand() * (gbest - particles[i])
            )

            particles[i] = particles[i] + velocities[i]

    return gbest

pso_weights = pso_optimize()

# =============================
# BAT ALGORITHM REFINEMENT
# =============================

def bat_refinement(initial_weights, bats=15, iterations=40):

    dim = len(initial_weights)

    bat_positions = np.random.rand(bats, dim) + initial_weights
    velocities = np.zeros((bats, dim))

    best_bat = initial_weights
    best_score = 0

    for _ in range(iterations):

        for i in range(bats):

            freq = np.random.rand()

            velocities[i] = velocities[i] + (bat_positions[i] - best_bat) * freq
            bat_positions[i] = bat_positions[i] + velocities[i]

            weights = bat_positions[i]

            pred = np.dot(X_train, weights)
            pred = (pred > np.median(pred)).astype(int)

            score = np.mean(pred == y_train)

            if score > best_score:
                best_score = score
                best_bat = weights

    return best_bat

best_weights = bat_refinement(pso_weights)

# =============================
# PREDICTION FUNCTION
# =============================

def hybrid_predict(X, weights):

    score = np.dot(X, weights)
    threshold = np.median(score)

    return (score > threshold).astype(int)

predictions = hybrid_predict(X_test, best_weights)

predictions = predictions % 3

accuracy = accuracy_score(y_test, predictions)

print("Hybrid PSO + BA Accuracy:", accuracy)

# =============================
# CONFUSION MATRIX
# =============================

cm = confusion_matrix(y_test, predictions)

plt.figure()
sns.heatmap(cm, annot=True)
plt.title("bso_confusion_matrix")
plt.savefig(OUTPUT_DIR + r"\bso_confusion_matrix.png")
plt.show()

# =============================
# HEATMAP
# =============================

plt.figure()

corr = df[[
"Coverage_Percentage",
"Gap_Households",
"ULB_Density"
]].corr()

sns.heatmap(corr, annot=True)

plt.title("bso_feature_heatmap")

plt.savefig(OUTPUT_DIR + r"\bso_heatmap.png")

plt.show()

# =============================
# PREDICTION GRAPH
# =============================

plt.figure()

plt.scatter(y_test, predictions)

plt.xlabel("Actual")
plt.ylabel("Predicted")

plt.title("bso_prediction_graph")

plt.savefig(OUTPUT_DIR + r"\bso_prediction_graph.png")

plt.show()

# =============================
# PRIORITY DISTRIBUTION
# =============================

plt.figure()

df["Priority"].value_counts().plot(kind="bar")

plt.title("bso_priority_distribution")

plt.savefig(OUTPUT_DIR + r"\bso_priority_distribution.png")

plt.show()

# =============================
# SAVE CSV RESULTS
# =============================

result_df = pd.DataFrame({
"Actual": y_test,
"Predicted": predictions
})

result_df.to_csv(
OUTPUT_DIR + r"\bso_predictions.csv",
index=False
)

df.to_csv(
OUTPUT_DIR + r"\bso_results.csv",
index=False
)

# =============================
# SAVE MODEL
# =============================

joblib.dump(best_weights,
OUTPUT_DIR + r"\bso_model.pkl")

with h5py.File(
OUTPUT_DIR + r"\bso_model.h5","w") as f:

    f.create_dataset(
    "weights",
    data=np.array(best_weights)
)

# =============================
# SAVE YAML
# =============================

metadata = {

"model": "Hybrid PSO + Bat Algorithm",

"features": [
"Coverage_Percentage",
"Gap_Households",
"ULB_Density"
],

"accuracy": float(accuracy)

}

with open(
OUTPUT_DIR + r"\bso_metadata.yaml","w"
) as f:

    yaml.dump(metadata,f)

# =============================
# SAVE JSON
# =============================

with open(
OUTPUT_DIR + r"\bso_results.json","w"
) as f:

    json.dump(result_df.to_dict(),f)

print("\nAll BSO hybrid results saved successfully")
```
