```python
import pandas as pd
import json
import yaml
import joblib
import h5py

from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.pipeline import Pipeline

# =============================
# FILE PATHS
# =============================

DATA_PATH = r"C:\Users\NXTWAVE\Downloads\Water Infrastructure Priority Prediction\RS_Session_259_AU_1152_C.csv"

OUTPUT_DIR = r"C:\Users\NXTWAVE\Downloads\Water Infrastructure Priority Prediction"

MODEL_PKL = OUTPUT_DIR + r"\water_priority_model.pkl"
MODEL_H5 = OUTPUT_DIR + r"\water_priority_model.h5"
YAML_FILE = OUTPUT_DIR + r"\model_metadata.yaml"
JSON_FILE = OUTPUT_DIR + r"\district_priority_predictions.json"

# =============================
# LOAD DATASET
# =============================

df = pd.read_csv(DATA_PATH)

print("\nDataset Loaded Successfully\n")
print(df.head())

# =============================
# RENAME COLUMNS (IMPORTANT)
# =============================

df = df.rename(columns={

"Name of District": "District",

"Number of Urban Local Bodies (ULBs) in the district":
"ULBs",

"Total households in the ULBs in 2021":
"Total_Households",

"Number of households in these ULBs with Tap connections":
"Tap_Connections",

"Number of households in these ULBs where Tap connections are yet to be provided":
"No_Tap"

})

# =============================
# FEATURE ENGINEERING
# =============================

df["Coverage_Percentage"] = (
    df["Tap_Connections"] /
    df["Total_Households"] * 100
)

df["Gap_Households"] = (
    df["Total_Households"] -
    df["Tap_Connections"]
)

df["ULB_Density"] = (
    df["ULBs"] /
    df["Total_Households"]
)

features = df[[
    "Coverage_Percentage",
    "Gap_Households",
    "ULB_Density"
]]

# =============================
# MACHINE LEARNING PIPELINE
# =============================

pipeline = Pipeline([
    ("scaler", StandardScaler()),
    ("kmeans", KMeans(n_clusters=3, random_state=42))
])

pipeline.fit(features)

clusters = pipeline.predict(features)

df["Cluster"] = clusters

# =============================
# CLUSTER → PRIORITY
# =============================

cluster_mean = df.groupby("Cluster")["Coverage_Percentage"].mean()

sorted_clusters = cluster_mean.sort_values()

priority_map = {}

priority_map[sorted_clusters.index[0]] = "High Priority"
priority_map[sorted_clusters.index[1]] = "Medium Priority"
priority_map[sorted_clusters.index[2]] = "Low Priority"

df["Priority"] = df["Cluster"].map(priority_map)

# =============================
# SAVE JSON PREDICTIONS
# =============================

predictions = df[[
    "District",
    "Coverage_Percentage",
    "Gap_Households",
    "Priority"
]].to_dict(orient="records")

with open(JSON_FILE, "w") as f:
    json.dump(predictions, f, indent=4)

print("\nJSON predictions saved")

# =============================
# SAVE MODEL PKL
# =============================

joblib.dump(pipeline, MODEL_PKL)

print("PKL model saved")

# =============================
# SAVE MODEL H5
# =============================

scaler = pipeline.named_steps["scaler"]
kmeans = pipeline.named_steps["kmeans"]

with h5py.File(MODEL_H5, "w") as f:

    f.create_dataset("scaler_mean", data=scaler.mean_)
    f.create_dataset("scaler_scale", data=scaler.scale_)
    f.create_dataset("cluster_centers", data=kmeans.cluster_centers_)

print("H5 model saved")

# =============================
# SAVE YAML METADATA
# =============================

metadata = {

"project": "Water Infrastructure Priority Prediction",

"model": "KMeans Clustering",

"features": [
"Coverage_Percentage",
"Gap_Households",
"ULB_Density"
],

"districts": df["District"].tolist(),

"dataset_size": int(len(df)),

"description":
"Districts are classified into High, Medium, and Low water infrastructure priority."

}

with open(YAML_FILE, "w") as f:
    yaml.dump(metadata, f)

print("YAML metadata saved")

# =============================
# SHOW FINAL RESULT
# =============================

print("\nFinal District Priority Table:\n")

print(df[[
"District",
"Coverage_Percentage",
"Gap_Households",
"Priority"
]])

print("\nPipeline Completed Successfully")
```
